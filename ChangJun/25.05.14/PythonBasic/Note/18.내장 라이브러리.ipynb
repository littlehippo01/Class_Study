```python
# OS (운영체제)
import os
# 현재 작업 디렉토리
cwd = os.getcwd()
print(cwd)

# 디렉토리 삭제
try :
    os.rmdir('example_dir')
except : 
    pass

# 파일 삭제
try :
    os.remove('example_file.txt')
except:
    pass

# 디렉토리 생성
os.makedirs('example_dir')

# 파일 생성
with open('example_file.txt','w') as f:
    f.write('Hello World!')

# 파일 이동
os.rename('example_file.txt', 'example_dir/example_file.txt')
```
